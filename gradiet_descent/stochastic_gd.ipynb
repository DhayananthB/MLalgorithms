```python
from sklearn.datasets import load_diabetes
import numpy as np 
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
X,y = load_diabetes(return_X_y=True)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
print("coeff : " , lr.coef_)
print("intercept: ",lr.intercept_)
print("r2_score : ",r2_score(y_test,lr.predict(X_test)))
```

```python
class SGDRegressor:
    def __init__(self,lr,epochs):
        self.lr=lr
        self.epochs=epochs
        self.m = None
        self.b = None
    def fit(self,X_train,y_train):
        self.b=0
        self.m = np.ones(X_train.shape[1])
        
        for i in range(self.epochs):
            for i in range(X_train.shape[0]):

                idx = np.random.randint(0,X_train.shape[0])

                y_pred = np.dot(X_train[idx],self.m)+ self.b

                

                b_der = -2*(y_train[idx]-y_pred)
                self.b -= self.lr*b_der

                m_der = -2*(y_train[idx]-y_pred)*X_train[idx]   
                self.m -= self.lr * m_der
            print(self.m,self.b)

    def predict(self,X):
        return np.dot(X,self.m)+self.b
```

```python
sgd = SGDRegressor(0.01,40)
```

```python
sgd.fit(X_train,y_train)
```

```python
r2_score(y_test,sgd.predict(X_test))
```
